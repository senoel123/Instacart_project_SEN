```python
#Define a function

def price_label(row):
    if row['prices'] <=5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'np.nan'
```

```python
#Apply a function

df['price_range'] = df.apply(price_label, axis=1)
```

### Check the values of new column for accuracy

By adding the values in result to a new column in our dataframe, we are fruther able to see how many sales occur on each of the different types of days.  This increases that scope of details that the end users (client) is asking for. 

```python
#Define a function

def busiest_period_of_day(row):
    if row['order_hour_of_day'] <10:
        return 'Fewest orders'
    elif (row['order_hour_of_day'] > 0) and (row['order_hour_of_day'] == 10):
        return 'Most orders'
    elif row['order_hour_of_day'] > 10:
        return 'Average orders'
    else: return 'np.nan'
```

```python
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
#If-Statements with the loc() 

ords_prods_merge.loc[ords_prods_merge['order_hour_of_day'] > 17, 'busiest_period_of_day'] = 'Average orders'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_hour_of_day'] <=17) & (ords_prods_merge['order_hour_of_day'] > 8), 'busiest_period_of_day'] = 'Highest orders'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_hour_of_day'] <= 8, 'busiest_period_of_day'] = 'Fewest orders'
```
