Create an order frequency flag that marks the regularity of a userâ€™s ordering behavior according to the median in the â€œdays_since_prior_orderâ€ column. 

```python
# convert string to an integer

ords_prods_merge['days_since_prior_order'] = ords_prods_merge['days_since_prior_order'].astype("Int64")
```

```python
print (ords_prods_merge['days_since_prior_order'].dtypes)
```

```python
ords_prods_merge['days_since_prior_order'] = ords_prods_merge['days_since_prior_order'].fillna(0)
```

```python
# Deriving Columns with loc()

ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'order_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] >10) & (ords_prods_merge['days_since_prior_order'] <= 20), 'order_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'order_flag'] = 'Frequent customer'
```
